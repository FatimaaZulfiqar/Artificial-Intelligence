```python
class Trapezoid:
    def __init__(self, a, b, h):
        self.a = a
        self.b = b
        self.h = h

    def area(self):
        return 0.5 * (self.a + self.b) * self.h


class Parallelogram:
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return self.base * self.height
```

```python
results = {}

data = [
    (10, 8, 5, 6, 7),
    (12, 10, 4, 5, 9),
    (15, 5, 6, 8, 3)
]

for i, (a, b, h, base, height) in enumerate(data):
    t = Trapezoid(a, b, h)
    p = Parallelogram(base, height)

    results[f"Set {i+1}"] = {
        "Trapezoid": t.area(),
        "Parallelogram": p.area()
    }

for key, value in results.items():
    print(key, value)

max_shape = None
max_value = 0

for dataset in results.values():
    for shape, area in dataset.items():
        if area > max_value:
            max_value = area
            max_shape = shape

print("Max Area Shape:", max_shape, "Area:", max_value)
```

```python
class Trapezoid:
    def __init__(self, a, b, h):
        self.a = a
        self.b = b
        self.h = h

    def area(self):
        return 0.5 * (self.a + self.b) * self.h


class Parallelogram:
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return self.base * self.height


# Create objects
t = Trapezoid(10, 8, 5)
p = Parallelogram(6, 7)

# Compare areas
if t.area() > p.area():
    print("Trapezoid has larger area")
elif t.area() < p.area():
    print("Parallelogram has larger area")
else:
    print("Both have equal area")
```
