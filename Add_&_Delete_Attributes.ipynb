```python
class Student:
    def __init__(self, student_id, student_name):
        self.student_id = student_id
        self.student_name = student_name

s = Student(1, "Fatima")

s.student_class = "BSE"

print("Before deletion:", s.__dict__)

del s.student_name

print("After deletion:", s.__dict__)
class Student:
    def __init__(self, student_id, student_name):
        self.student_id = student_id
        self.student_name = student_name

s = Student(1, "Fatima")

s.student_class = "BSE"

print("Before deletion:", s.__dict__)

del s.student_name

print("After deletion:", s.__dict__)
```
