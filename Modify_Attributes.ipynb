```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

s = Student("Fatima", 85)

print("Original:", s.name, s.marks)

s.name = "Zara"
s.marks = 90

print("Updated:", s.name, s.marks)
```
