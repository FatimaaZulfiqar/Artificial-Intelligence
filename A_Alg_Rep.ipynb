```python
import heapq

# A* Algorithm function
def a_star(graph, heuristics, start, goal):

    # Priority queue (min-heap)
    open_list = []
    heapq.heappush(open_list, (0, start))

    # To track visited nodes
    came_from = {}

    # Cost from start to each node
    g_cost = {node: float('inf') for node in graph}
    g_cost[start] = 0

    # Estimated total cost
    f_cost = {node: float('inf') for node in graph}
    f_cost[start] = heuristics[start]

    while open_list:

        current_f, current = heapq.heappop(open_list)

        # If goal reached, reconstruct path
        if current == goal:
            path = []
            while current in came_from:
                path.append(current)
                current = came_from[current]
            path.append(start)
            return path[::-1]

        # Explore neighbors
        for neighbor, cost in graph[current].items():

            tentative_g = g_cost[current] + cost

            if tentative_g < g_cost[neighbor]:
                came_from[neighbor] = current
                g_cost[neighbor] = tentative_g
                f_cost[neighbor] = tentative_g + heuristics[neighbor]

                heapq.heappush(open_list, (f_cost[neighbor], neighbor))

    return None


# ------------------ Example Graph ------------------

graph = {
    'A': {'B': 1, 'C': 4},
    'B': {'A': 1, 'D': 2, 'E': 5},
    'C': {'A': 4, 'F': 3},
    'D': {'B': 2},
    'E': {'B': 5, 'F': 1},
    'F': {'C': 3, 'E': 1}
}

# Heuristic values (estimated cost to goal F)
heuristics = {
    'A': 7,
    'B': 6,
    'C': 2,
    'D': 6,
    'E': 1,
    'F': 0
}

start = 'A'
goal = 'F'

path = a_star(graph, heuristics, start, goal)

print("Shortest Path using A*:", path)
```
